```python
temp = 0
for c in range (0,20):
    for i in range (0,16):
        j = 0
        prod1 = 1
        prod2 = 1
        while j < 4:
            prod1 *= table.iat[i+j,c]
            prod2 *= table.iat[c,i+j]
            j += 1
        prod = max([prod1,prod2])
        """ print(prod) """
        if temp < prod:
            temp = prod
print(temp)
    
```
